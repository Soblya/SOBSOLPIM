```python
# импортируем библиотеки
import tensorflow as tf
from tensorflow import keras
from keras.preprocessing.image import ImageDataGenerator
import seaborn
import matplotlib
from keras.layers import Activation, AveragePooling2D, BatchNormalization, Conv2D, Dense, Dropout, DepthwiseConv2D, Flatten, InputLayer, LeakyReLU, MaxPool2D, MaxPooling2D
from keras.losses import BinaryCrossentropy, CategoricalCrossentropy
from keras.optimizers import Adam
import keras.callbacks
```

```python
image_gen = ImageDataGenerator(rescale=1./255,
                               validation_split = 0.22, # 22% данных оставляем на валидацию
                               zoom_range=(0.99,0.99),
                               dtype=tf.float32)
```

```python
training_ds = image_gen.flow_from_directory("./dataset/",
                                            target_size = (150,150),
                                            batch_size = 256,
                                            class_mode = "binary",
                                            color_mode = "rgb",
                                            shuffle = True,
                                            seed = 3347,
                                            subset = "training")
```

```python
validation_ds = image_gen.flow_from_directory("./dataset/",
                                              target_size = (150,150),
                                              batch_size = 256,
                                              class_mode = "binary",
                                              color_mode = "rgb",
                                              shuffle = True,
                                              seed = 3347,
                                              subset = "validation")
```

Тут видно, что есть небольшой перевес данных в датасете в сторону больных

```python
# базовая модель
mdl = keras.Sequential() # группа последовательных слоёв
mdl.add(InputLayer(input_shape=(150,150,3))) # вводной слой
mdl.add(Conv2D(filters=32, kernel_size=3, activation='relu', padding='same')) # サメにおつき路
mdl.add(MaxPool2D())
mdl.add(Conv2D(filters=64, kernel_size=3, activation='relu', padding='same'))
mdl.add(MaxPool2D())

mdl.add(Flatten())

mdl.add(Dense(128, activation='relu')) # Плотно связаный слой размерностью 128
mdl.add(BatchNormalization()) # Слой нормализации
#mdl.add(Dropout(rate=0.25))
mdl.add(Dense(128, activation='relu')) # Плотно связаный слой размерностью 128
mdl.add(BatchNormalization()) # Слой нормализации
mdl.add(Dropout(rate=0.25)) # Слой отбрасывания, отбрасывает 25% случайных нейронов, 
                            # во избежание сложных коадаптаций и, как следствие переобучения
mdl.add(Dense(64, activation='relu')) # Плотно связаный слой размерностью 64
mdl.add(BatchNormalization()) # Слой нормализации
mdl.add(Dropout(rate=0.25))
mdl.add(Dense(1, activation='sigmoid')) # Плотно связаный слой размерностью 1

mdl.compile(optimizer=Adam(0.001), loss=BinaryCrossentropy(), metrics=['accuracy']) # компиляция модели,метрика обучения -- 
                                                                                    # точность предсказания
mdl.summary() # Текстовое описание свёрток
```

Визуализация модели

```python
from tensorflow.keras.utils import plot_model
from IPython.display import Image
plot_model(mdl, to_file='convnet.png', show_shapes=True,show_layer_names=True)
Image(filename='convnet.png')
```

Базовая модель. 
После обучения (15 поколений) -- имеем следующие рез-ты: 

* Синий: точность на тренировочном датасете
* Оранжевый: точность на валидационном датасете



```python
# хорошая, в общем-то, модель. Тут как метрика дополнительно к точности используются FalseNegatives, ибо сеть,
# вычисляющая шанс наличия заболевания, должна минимизировать количество ложноотрицательных срабатываний.
# Лучше дополнительно обследовать здорового, чем посчитать больного здоровым
mdl = keras.Sequential()
mdl.add(InputLayer(input_shape=(150,150,3)))
mdl.add(Conv2D(filters=32, kernel_size=3, activation='relu', padding='same'))
mdl.add(MaxPool2D())
mdl.add(Conv2D(filters=64, kernel_size=3, activation='relu', padding='same'))
mdl.add(MaxPool2D())

mdl.add(Flatten())

mdl.add(Dense(128, activation='relu'))
mdl.add(BatchNormalization())
mdl.add(Dropout(rate=0.3))
mdl.add(Dense(64, activation='relu'))
mdl.add(BatchNormalization())
mdl.add(Dropout(rate=0.3))
mdl.add(Dense(1, activation='sigmoid'))

mdl.compile(optimizer=Adam(0.001), loss=BinaryCrossentropy(), metrics=['accuracy', 'MeanSquaredError', 'FalseNegatives'])
mdl.summary()
```

```python
from tensorflow.keras.utils import plot_model
from IPython.display import Image
plot_model(mdl, to_file='convnet.png', show_shapes=True,show_layer_names=True)
Image(filename='convnet.png')
```

После 15 эпох обучения имеем следующие результаты.

* Синий: точность на тренировочном датасете
* Оранжевый: точность на валидационном датасете
* Зеленый: среднеквадратичное отклонение на тренировочном датасете
* Красный: среднеквадратичное отклонение на валидационном датасете



```python
import math
import datetime

def step_decay(epoch):
    initial_lrate=0.1
    drop=0.6
    epochs_drop = 3.0
    lrate= initial_lrate * math.pow(drop,  
           math.floor((1+epoch)/epochs_drop))
    return lrate
   

lrate = tf.keras.callbacks.LearningRateScheduler(step_decay)

callbacks_list = [lrate]
```

```python
history = mdl.fit(training_ds, verbose=1, epochs=10, callbacks=callbacks_list, validation_data=(validation_ds))
```

```python
matplotlib.pyplot.plot(history.history['accuracy'], label='acc')
matplotlib.pyplot.plot(history.history['val_accuracy'], label='val_acc')
matplotlib.pyplot.plot(history.history['mean_squared_error'], label='MSR')
matplotlib.pyplot.plot(history.history['val_mean_squared_error'], label='val_MSR')

matplotlib.pyplot.xlabel("Ep")
matplotlib.pyplot.ylabel("Acc")
```

```python
from keras.models import Model
layer_outputs = [layer.output for layer in mdl.layers]
activation_model = Model(inputs=mdl.input, outputs=layer_outputs)
activations = activation_model.predict(training_ds[0][0])
 
def display_activation(activations, col_size, row_size, act_index): 
    activation = activations[act_index]
    activation_index=0
    fig, ax = matplotlib.pyplot.subplots(row_size, col_size, figsize=(row_size*2.5,col_size*1.5))
    for row in range(0,row_size):
        for col in range(0,col_size):
            ax[row][col].imshow(activation[1, :, :, activation_index], cmap='gray')
            activation_index += 1
```
